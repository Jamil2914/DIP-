# Prewitt Operator Vertical

```python
img = cv.imread(path,0)
```

```python
img.shape
```

```python
no_of_row = img.shape[0]
```

```python
no_of_column= img.shape[1]
```

```python
no_of_row
```

```python
no_of_column
```

```python
k = np.array([[-1,0,1],
             [-1,0,1],
             [-1,0,1]])
```

```python
k.shape[0]
```

```python
padded = np.zeros((no_of_row + k.shape[0]-1, no_of_column + k.shape[1]-1))

```

```python
Output_Image = np.zeros(img.shape)

```

```python
padded[1:-1, 1:-1] = img
```

```python
for x in range(no_of_row): 
    for y in range(no_of_column):
        Output_Image[x, y] = (k * padded[x:x + 3,y:y + 3]).sum()
```

```python
plt.figure(figsize=(20,12))
plt.subplot(3,3,1)
plt.imshow(Output_Image)
```

# Prewitt Operator Horizontal

```python
k = np.array([[-1,-1,1],
             [0,0,0],
             [1,0,1]])
```

```python
img = cv.imread(path,0)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,1,1)
plt.imshow(img,cmap='gray')
```

```python
padded = np.zeros((no_of_row + k.shape[0]-1, no_of_column + k.shape[1]-1))
PrewittHor = np.zeros(img.shape)
padded[1:-1, 1:-1] = img
```

```python
for x in range(no_of_row): 
    for y in range(no_of_column):
        PrewittHor[x, y] = (k * padded[x:x + 3,y:y + 3]).sum()
```

```python
plt.figure(figsize=(20,12))
plt.subplot(3,3,1)
plt.imshow(PrewittHor,cmap='gray')
```
